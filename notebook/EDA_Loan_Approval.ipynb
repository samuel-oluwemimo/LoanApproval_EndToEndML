# EDA
### Applicant Demographics and Profiles

```python
#What is the age distribution of applicants?
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age')
plt.show()
#observation: Older people don't really apply for loans
#                middle aged people tend to apply more
```

```python
#What are the income levels and their relationship with employment status or education level?

agg_data = df.groupby(["EmploymentStatus", "EducationLevel"])["MonthlyIncome"].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x="EmploymentStatus", y="MonthlyIncome", hue="EducationLevel", data=agg_data)
plt.title("Income Levels by Employment Status and Education Level")
plt.ylabel("Average Monthly Income")
plt.xticks(rotation=45)
plt.show()

"""Observation: on average in the employed category, a doctorate degree makes more than an high school degree,
                while a bachelor degree makes more than a doctorate and master's degree unemployed"""
```

```python
#How does marital status correlate with the number of dependents?
df.groupby('MaritalStatus')['NumberOfDependents'].mean().reset_index()
```

```python
sns.barplot(x='MaritalStatus', y='NumberOfDependents', data=df)
plt.show()
#Observation: there isn't that much of a difference but married people have more dependents on avg with widowed being the lowest
```

```python
#What are the common credit scores among applicants?
creditCount = df.groupby('CreditScore')['CreditScore'].count()
creditCount.sort_values(ascending=False)[:10] # top 10 most common credit scores
```

### Loan Characteristics

```python
#What are the most common loan purposes?
plt.figure(figsize=(10, 6))
sns.histplot(x='LoanPurpose', data=df)
plt.show()
#Observation: The common loan purpose is Home->Debt->Auto->Education->other
```

```python
#How does loan amount vary with annual income and credit score?
# Bin data for heatmap
df['CreditScoreBins'] = pd.cut(df['CreditScore'], bins=10)
df['IncomeBins'] = pd.cut(df['AnnualIncome'], bins=10)

# Pivot table for heatmap
heatmap_data = df.pivot_table(
    index='IncomeBins', columns='CreditScoreBins', values='LoanAmount', aggfunc='mean'
)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".1f")
plt.title('Loan Amount by Credit Score and Annual Income')
plt.xlabel('Credit Score Bins')
plt.ylabel('Annual Income Bins')
plt.show()
```

```python
#Are shorter loan durations associated with smaller loan amounts?
sns.boxplot(x='LoanDuration', y='LoanAmount', data=df)
plt.show()


"""
Observation:
Based on the boxplot, it appears there is no significant association between shorter
 loan durations and smaller loan amounts. The median loan amount across different loan
   durations remains relatively constant, and there is no clear trend of lower loan amounts
     for shorter durations. While there might be some variation in the spread or outliers, 
     the overall distribution of loan amounts does not strongly depend on loan duration.
     """
```

```python
correlation = df['LoanDuration'].corr(df['LoanAmount'])
print(f"Correlation: {correlation}")
```

### Financial Behavior

```python
#Do applicants with bankruptcy history request higher loan amounts?
agg = df.groupby('BankruptcyHistory')['LoanAmount'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='BankruptcyHistory', y='LoanAmount', data=agg)
plt.show()
```

```python
from scipy.stats import ttest_ind

group_1 = df[df['BankruptcyHistory'] == 0]['LoanAmount']
group_2 = df[df['BankruptcyHistory'] == 1]['LoanAmount']

t_stat, p_value = ttest_ind(group_1, group_2, equal_var=False)
print(f"T-Statistic: {t_stat}, P-Value: {p_value}")
# it means the difference in the loan amounts requested by applicants with and without a bankruptcy history is not statistically significant
# BankruptcyHistory is not a good metric for determining loan amount
```

```python
#What percentage of applicants have a high risk score?
threshold = 54

highRiskScore = df[df["RiskScore"]  > threshold]['RiskScore'].count()
totalApplicants = df['RiskScore'].count()
print(f"Percentange of applicants with high risk scores are {(highRiskScore/totalApplicants) *100}%")
```

```python
#How does debt-to-income ratio vary across different income levels?
sns.scatterplot(x='TotalDebtToIncomeRatio', y='MonthlyIncome', data=df)
plt.show()
```

### Loan Approval and Risk

```python
#What factors most influence loan approval (e.g., CreditScore, LoanPurpose, DebtToIncomeRatio)?
df = df.select_dtypes(include=['number'])
correlation = df.corr()['LoanApproved'].sort_values(ascending=False)
print(correlation)
```

```python
#How do payment history and previous loan defaults correlate with loan approval?

fig, axs = plt.subplots(1, 2, figsize=(13, 5))
plt.subplot(121)
sns.histplot(data=df,x='PaymentHistory',bins=30,kde=True,color='g',hue='LoanApproved')
plt.subplot(122)
sns.histplot(data=df,x='PreviousLoanDefaults',hue='LoanApproved')
plt.show()
```
